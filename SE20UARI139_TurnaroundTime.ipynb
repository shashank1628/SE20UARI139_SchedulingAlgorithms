```python
def calculate_waiting_turnaround(processes, algorithm, time_quantum=None):
    n = len(processes)
    waiting_time = [0] * n
    turnaround_time = [0] * n
    remaining_time = [process[2] for process in processes]
    current_time = 0

    while sum(remaining_time) > 0:
        if algorithm == "FCFS":
            next_process = None
            for i in range(n):
                if int(processes[i][1]) <= current_time and remaining_time[i] > 0:
                    next_process = i
                    break
        elif algorithm == "SJF":
            next_process = None
            min_burst_time = float('inf')
            for i in range(n):
                if int(processes[i][1]) <= current_time and remaining_time[i] > 0 and remaining_time[i] < min_burst_time:
                    min_burst_time = remaining_time[i]
                    next_process = i
        elif algorithm == "PS":
            next_process = None
            min_priority = float('inf')
            for i in range(n):
                if int(processes[i][1]) <= current_time and remaining_time[i] > 0 and int(processes[i][3]) < min_priority:
                    min_priority = int(processes[i][3])
                    next_process = i
        elif algorithm == "RR":
            next_process = None
            for i in range(n):
                if int(processes[i][1]) <= current_time and remaining_time[i] > 0:
                    next_process = i
                    break

        if next_process is None:
            current_time += 1
        else:
            if algorithm == "RR" and remaining_time[next_process] > time_quantum:
                remaining_time[next_process] -= time_quantum
                current_time += time_quantum
            else:
                waiting_time[next_process] = current_time - int(processes[next_process][1])
                current_time += remaining_time[next_process]
                turnaround_time[next_process] = waiting_time[next_process] + int(processes[next_process][2])
                remaining_time[next_process] = 0

    return waiting_time, turnaround_time

# Process Data: (name, arrival time, burst time, priority)
processes = [
    ('P1', 0, 24, 3),
    ('P2', 4, 3, 1),
    ('P3', 5, 3, 4),
    ('P4', 6, 12, 2)
]

# FCFS
fcfs_waiting_time, fcfs_turnaround_time = calculate_waiting_turnaround(processes, "FCFS")

# Sort for SJF (Shortest Job First)
sjf_sorted_processes = sorted(processes, key=lambda x: x[2])
sjf_waiting_time, sjf_turnaround_time = calculate_waiting_turnaround(sjf_sorted_processes, "SJF")

# Sort for PS (Priority Scheduling)
ps_sorted_processes = sorted(processes, key=lambda x: x[3])
ps_waiting_time, ps_turnaround_time = calculate_waiting_turnaround(ps_sorted_processes, "PS")

# RR (Round Robin with a Time Quantum of 4 Units)
time_quantum = 4
rr_waiting_time, rr_turnaround_time = calculate_waiting_turnaround(processes, "RR", time_quantum)

# Calculate average waiting time and average turnaround time
def calculate_average(times):
    return sum(times) / len(times)

avg_waiting_fcfs = calculate_average(fcfs_waiting_time)
avg_turnaround_fcfs = calculate_average(fcfs_turnaround_time)
avg_waiting_sjf = calculate_average(sjf_waiting_time)
avg_turnaround_sjf = calculate_average(sjf_turnaround_time)
avg_waiting_ps = calculate_average(ps_waiting_time)
avg_turnaround_ps = calculate_average(ps_turnaround_time)
avg_waiting_rr = calculate_average(rr_waiting_time)
avg_turnaround_rr = calculate_average(rr_turnaround_time)

# Print results
print("FCFS Waiting Time:", fcfs_waiting_time)
print("FCFS Turnaround Time:", fcfs_turnaround_time)
print("Average FCFS Waiting Time:", avg_waiting_fcfs)
print("Average FCFS Turnaround Time:", avg_turnaround_fcfs)

print("\nSJF Waiting Time:", sjf_waiting_time)
print("SJF Turnaround Time:", sjf_turnaround_time)
print("Average SJF Waiting Time:", avg_waiting_sjf)
print("Average SJF Turnaround Time:", avg_turnaround_sjf)

print("\nPS Waiting Time:", ps_waiting_time)
print("PS Turnaround Time:", ps_turnaround_time)
print("Average PS Waiting Time:", avg_waiting_ps)
print("Average PS Turnaround Time:", avg_turnaround_ps)

print("\nRR Waiting Time:", rr_waiting_time)
print("RR Turnaround Time:", rr_turnaround_time)
print("Average RR Waiting Time:", avg_waiting_rr)
print("Average RR Turnaround Time:", avg_turnaround_rr)
```
